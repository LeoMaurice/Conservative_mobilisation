```python
liste_url1 =[]
liste_url2 = []
def extract_article_links(url):
    page = requests.get(url)
    time.sleep(1)  # Ajoutez une pause d'1 seconde entre les requêtes
    soup = BeautifulSoup(page.content, "html.parser")
    article_links = soup.find_all("a", class_="image-wrapper")  # Remplacez par la classe réelle
    return [link.get("href") for link in article_links if link.get("href")]

# URL des deux pages
page1_url = "https://www.femelliste.com/articles-femellisme-feminisme"
page2_url = "https://www.femelliste.com/articles-femellisme-feminisme?offset=1666117260433"

# Extraire les liens d'article de la première page
article_links_page1 = extract_article_links(page1_url)

# Afficher les liens d'article de la première page
for article_url in article_links_page1:
    full_article_url = page1_url + article_url if not article_url.startswith("http") else article_url
    liste_url1.append(full_article_url)
    print(f"URL de l'article (Page 1) : {full_article_url}")

# Extraire les liens d'article de la deuxième page
article_links_page2 = extract_article_links(page2_url)

# Afficher les liens d'article de la deuxième page
for article_url in article_links_page2:
    full_article_url = page2_url + article_url if not article_url.startswith("http") else article_url
    liste_url2.append(full_article_url)
    print(f"URL de l'article (Page 2) : {full_article_url}")
```

```python
def extract_title(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    titles = soup.find_all("h1")
    return titles
    
t1 = extract_title(page1_url)
t2 = extract_title(page2_url)
```

```python
t1l=[]
t2l=[]

for titre in t1:
        titre = (titre.text.strip()) 
        t1l.append(titre)

for titre in t2:
        titre = (titre.text.strip()) 
        t2l.append(titre)
titres_art = t1l + t2l
```

```python

# Créer un ensemble pour stocker les URL uniques
ensemble_urls = []

for article_url in article_links_page1:
    full_article_url = urljoin(page1_url, article_url)
    ensemble_urls.append(full_article_url)

for article_url in article_links_page2:
    full_article_url = urljoin(page1_url, article_url)
    ensemble_urls.append(full_article_url)

# Afficher les URL uniques
for unique_url in ensemble_urls:
    print(f"URL de l'article (Page 1) : {unique_url}")
```

```python
# Créer un dictionnaire avec les deux listes
data = {'url': ensemble_urls, 'titre_article': titres_art}

# Créer une DataFrame à partir du dictionnaire
df = pd.DataFrame(data)
df = df[['titre_article', 'url']]
df
```

```python
df.drop('contenu', axis=1)
```

```python
def get_page_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Trouver la balise d'article avec une classe ou un sélecteur CSS plus spécifique
    article_content_tag = soup.find('div', class_="sqs-html-content")

    # Extraire le texte du contenu de l'article
    contenu = article_content_tag.get_text(separator=' ', strip=True) if article_content_tag else None

    return contenu

df['contenu'] = df['url'].apply(get_page_content)
df
```
